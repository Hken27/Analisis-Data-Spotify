# **Load Dataset**

```python
import pandas as pd
import gdown

drive_url = "https://drive.google.com/file/d/1VL2loELaYxLPNlls2JaX1q4P1vNhbhmq/view?usp=sharing"

# Konversi URL berbagi ke format unduh langsung
file_id = drive_url.split("/d/")[1].split("/view")[0]
download_url = f"https://drive.google.com/uc?id={file_id}"

# Unduh file menggunakan gdown
output = "spotify_songs.csv"  # Nama file yang diunduh
gdown.download(download_url, output, quiet=False)

dataset = pd.read_csv(output)
```

# **EDA**

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = dataset.select_dtypes(include=['number']).columns

# Histograms for numeric columns
for col in numeric_cols:
  plt.figure(figsize=(8, 6))
  sns.histplot(dataset[col], kde=True)
  plt.title(f'Distribution of {col}')
  plt.show()
```

# **Analisa Data**

```python
# Mengganti nilai NaN dengan "Unknown" pada kolom string
dataset.fillna({
    "track_name": "Unknown",
    "track_artist": "Unknown",
    "track_album_name": "Unknown"
}, inplace=True)

# Menambahkan kolom baru: tempo_category berdasarkan nilai tempo
def categorize_tempo(tempo):
    if tempo < 90:
        return "Slow"
    elif 90 <= tempo <= 120:
        return "Moderate"
    else:
        return "Fast"

dataset["tempo_category"] = dataset["tempo"].apply(categorize_tempo)

# Menambahkan kolom baru: energy_acoustic_ratio
dataset["energy_acoustic_ratio"] = dataset["energy"] / (dataset["acousticness"] + 1e-6)

# Menambahkan kolom baru: energy_efficiency
dataset["energy_efficiency"] = dataset["loudness"] / dataset["tempo"]
```

```python
unknown_track_album_names = dataset[dataset['track_name'] == "Unknown"]
unknown_track_album_names
```

```python
# Mengecek hasil awal untuk kolom baru
dataset[["track_name", "tempo_category", "energy_acoustic_ratio", "energy_efficiency"]].head()
```

```python
# menghilangkan angka, simbol, dan karakter khusus
dataset["track_name"] = dataset["track_name"].str.replace(r"[^a-zA-Z0-9\s]", "", regex=True)
```

```python
avg_popularity = dataset['track_popularity'].mean()

popular_songs = dataset[dataset['track_popularity'] > avg_popularity]
popular_songs[['track_name', 'track_artist', 'track_popularity']].head()
```

```python
# Top genres by track popularity (averaged)
top_genres = dataset.groupby('playlist_genre')['track_popularity'].mean().sort_values(ascending=False)

# Plot 1: Top genres by popularity
plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='mako')
plt.title('Average Track Popularity by Genre', fontsize=16, weight='bold')
plt.xlabel('Average Popularity', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.show()
```

```python
# Mengatur gaya visualisasi
sns.set(style="whitegrid")
```

```python
# Grafik 1: Distribusi tempo_category
plt.figure(figsize=(8, 5))
sns.countplot(x="tempo_category", data=dataset, palette="pastel")
plt.title("Distribusi Tempo Category", fontsize=14, weight='bold')
plt.xlabel("Tempo Category", fontsize=12)
plt.ylabel("Jumlah Lagu", fontsize=12)
plt.show()

# Grafik 2: Energy Efficiency per Tempo Category
plt.figure(figsize=(10, 6))
sns.boxplot(x="tempo_category", y="energy_efficiency", data=dataset, palette="coolwarm")
plt.title("Energy Efficiency per Tempo Category", fontsize=14, weight='bold')
plt.xlabel("Tempo Category", fontsize=12)
plt.ylabel("Energy Efficiency", fontsize=12)
plt.show()

# Tabel 1: Ringkasan Statistik Berdasarkan Tempo Category
tempo_summary = dataset.groupby("tempo_category")[
    ["energy", "acousticness", "loudness", "energy_efficiency"]
].mean().reset_index()

tempo_summary.rename(
    columns={
        "energy": "Rata-rata Energy",
        "acousticness": "Rata-rata Acousticness",
        "loudness": "Rata-rata Loudness",
        "energy_efficiency": "Rata-rata Energy Efficiency",
    },
    inplace=True,
)

tempo_summary
```

```python
dataset['popularity_score'] = (
    dataset['track_popularity'] +
    dataset['danceability'] +
    dataset['energy'] -
    dataset['speechiness']
) / 4

# Plot 1: Hubungan antara Popularity Score dan Energy
plt.figure(figsize=(10, 6))
sns.regplot(
    x='energy',
    y='popularity_score',
    data=dataset,
    scatter_kws={'alpha': 0.5},
    line_kws={'color': 'red'},
    color='blue'
)
plt.title('Hubungan Popularity Score dan Energy', fontsize=16, weight='bold')
plt.xlabel('Energy', fontsize=12)
plt.ylabel('Popularity Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Plot 2: Distribusi Popularity Score berdasarkan Genre
plt.figure(figsize=(12, 6))
sns.boxplot(
    x='playlist_genre',
    y='popularity_score',
    data=dataset,
    palette='coolwarm'
)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Memutar label sumbu x
plt.title('Distribusi Popularity Score berdasarkan Genre', fontsize=16, weight='bold')
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Popularity Score', fontsize=12)
plt.tight_layout()  # Menyesuaikan tata letak agar tidak ada overlap
plt.show()
```

```python
# top 10 lagu
top_songs = dataset.sort_values(by='popularity_score', ascending=False).head(10)
top_songs[['track_name', 'track_artist', 'track_popularity', 'popularity_score', 'playlist_genre']]

```
